## 🎯 動機：好內容 ≠ 你要的內容

你有沒有用過 Google 的 **NotebookLM**？你只要丟資料進去，它就能在 30 秒內自動生成：
- 🎙️ 雙人對話 podcast（男聲 + 女聲，自然得像真人）

### 但問題來了：NotebookLM 是黑箱

NotebookLM 生成的內容品質確實不錯，但 **「不錯」不等於「你想要的」**。

你無法控制它如何解讀你的資料、如何組織敘事、如何選擇重點。它可能會：
- 把你不重要的段落當成核心主題
- 忽略你真正想傳達的關鍵資訊
- 用你不喜歡的語氣或角度來呈現內容

換句話說：**NotebookLM 給你「好內容」，但不保證給你「你要的內容」。**

### LangChain：把控制權拿回來

這就是為什麼我們需要學習 LangChain。

在本課程中，你將學會：
- 用 **Prompt 設計** 精準控制 LLM 的語氣、結構與重點
- 用 **RAG 知識庫** 從龐大資料中擷取你真正需要的段落
- 用 **Chain 與 Agent** 打造自動化的內容處理流程
- 最後，把「你精心調校過的內容」餵給 NotebookLM

這樣一來，NotebookLM 生成的不再是隨機的「不錯內容」，而是 **真正符合你需求的內容**。

本 Notebook 將先帶你體驗 NotebookLM 的強大功能，讓你親眼看到「沒有控制」的 AI 能做出什麼。接下來的 8 週，我們將一步步學會如何把控制權拿回來。

## 📓 建立 Notebook 與加入來源

NotebookLM 的核心概念是 **Notebook**（筆記本），每個筆記本可以加入多個 **來源（Sources）**：
- 網頁 URL
- PDF 檔案
- 純文字
- YouTube 影片
- 音檔

API:
    - add_url
    - add_text
    - add_drive
    - add_file

來源加入後，NotebookLM 會自動解析內容，之後你就可以對它提問或生成各種 artifact。


````python
async with await NotebookLMClient.from_storage(timeout=600) as client:
    # 建立一個新的 Notebook
    nb = await client.notebooks.create("Warhammer 40k Lore")
    print(f"📓 Created notebook: {nb.id}")
    
    # 加入來源：Wikipedia 頁面
    await client.sources.add_url(
        nb.id,
        "https://zh.wikipedia.org/zh-tw/%E6%88%98%E9%94%A440000",
        wait=True  # 等待來源解析完成
    )
    print("✅ Source added and processed")
````

加入來源後，你可以直接對筆記本提問，NotebookLM 會基於你提供的資料回答（本質上就是 Google 版的 RAG）。


````python
async with await NotebookLMClient.from_storage(timeout=600) as client:
    # 注意：這裡的 nb 沿用上一個 cell 建立的筆記本
    # 在實際使用中，你可以透過 nb.id 或 client.notebooks.list() 找到已建立的筆記本
    
    result = await client.chat.ask(nb.id, "Summarize the Warhammer 40k universe in 3 sentences")
    print(result.answer)
````

## 🎙️ 生成 Podcast（Audio Overview）

這是 NotebookLM 最有名的功能：**雙人對話 podcast**。

兩個 AI 主持人（一男一女）會用極其自然的對話方式討論你的資料，笑聲、語氣、互動都非常逼真。

### 參數說明

| 參數 | 說明 | 可選值 |
|------|------|--------|
| `instructions` | 自訂生成指令 | 任意文字，如 "make it fun", "focus on the dark themes" |
| `audio_length` | 音檔長度 | `SHORT` (~3-5 分鐘), `LONG` (~10-20 分鐘) |
| `audio_format` | 音檔格式 | `DEEP_DIVE`, `BRIEF`, `CRITIQUE`, or `DEBATE`|


````python
from notebooklm.rpc.types import AudioLength, AudioFormat

async with await NotebookLMClient.from_storage(timeout=600) as client:
    # 觸發 podcast 生成
    status = await client.artifacts.generate_audio(
        nb.id,
        instructions="用幽默風趣的方式，介紹斯萊馬博。全程使用中文",
        audio_length=AudioLength.SHORT,
        audio_format=AudioFormat.BRIEF,
        language='zh_Hant'
    )
    print(f"🎙️ Generation started: {status.task_id}")
````

````python
# 不斷地檢查任務是否完成
while True:

    async with await NotebookLMClient.from_storage(timeout=600) as client:
    
        is_done = await client.artifacts.poll_status(nb.id, task_id=status.task_id)

        if is_done.status=='completed':
            await client.artifacts.download_audio(nb.id, "sly_marbo_podcast.mp3")
            print("📥 Downloaded: sly_marbo_podcast.mp3")
            # 任務完成，離開
            break
        else:
            # 還沒完成就等10秒
            asyncio.sleep(10)
````

### 參數說明

| 參數 | 說明 | 可選值 |

| `detail_level` | 細節程度 | `CONCISE` or `STANDARD` or `DETAILED`|
| `style` | 風格 | `ANIME` or `KAWAII` or `BENTO_GRID`...etc|

````python
import asyncio
from notebooklm import NotebookLMClient
````

````python
# 初始化 client（自動讀取本地儲存的 credentials）
client = await NotebookLMClient.from_storage(timeout=600)
print("✅ Client initialized")
````

````python
algorithm_steps= [
        "1. Consider object moving at constant speed v in circle radius r",
        "2. Draw velocity vectors at two nearby points - same magnitude, different direction",
        "3. The change in velocity Δv points toward the center",
        "4. For small angle Δθ: |Δv| ≈ vΔθ (isosceles triangle approximation)",
        "5. Substitute Δθ = ωΔt = (v/r)Δt",
        "6. Get |Δv| ≈ (v²/r)Δt",
        "7. Acceleration a = |Δv|/Δt = v²/r toward center"
      ]

content = "\n\n".join(algorithm_steps)
````

````python
async with await NotebookLMClient.from_storage(timeout=600) as client:
    # 建立一個新的 Notebook
    nb = await client.notebooks.create("Test")
    print(f"📓 Created notebook: {nb.id}")
    
    await client.sources.add_text(
        nb.id,
        content=content,
        title="Deriving Centripetal Acceleration",
        wait=True  # 等待來源解析完成
    )
    print("✅ Source added and processed")
````

## 🔄 非同步背景任務

生成 artifact 可能需要幾分鐘。你可以用 `asyncio.create_task()` 在背景執行，同時做其他事情。


````python
async def generate_slides_background(client, nb_id):
    print("🔄 Background task: Generating slides...")
    status = await client.artifacts.generate_slide_deck(
        nb_id,
        instructions="make it fun",
        slide_format=SlideDeckFormat.PRESENTER_SLIDES,
        slide_length=SlideDeckLength.SHORT
    )
    await client.artifacts.wait_for_completion(nb_id, status.task_id)
    await client.artifacts.download_slide_deck(nb_id, "slide_deck_bg.pdf")
    print("✅ Background task complete: slide_deck_bg.pdf")


async with await NotebookLMClient.from_storage(timeout=600) as client:
    # 建立新筆記本
    nb = await client.notebooks.create("Titan Lore")
    await client.sources.add_url(
        nb.id,
        "https://warhammer40k.fandom.com/wiki/Titan",
        wait=True
    )
    
    # 啟動背景任務
    task = asyncio.create_task(generate_slides_background(client, nb.id))
    
    # 同時做其他事情
    for i in range(5):
        print(f"📌 Doing other work... {i}")
        await asyncio.sleep(1)
    
    print("✅ Main work done")
````
